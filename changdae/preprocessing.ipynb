# 케이스 n 합치기 - 함수정의

```python
def function_concat(num):
    case_temp = []
    for i in range(1,6):
        case_temp.append(pd.read_csv(f'semina_experiment/{num}-{i}.csv', header=None)[5:])

    case = pd.concat(case_temp).reset_index(drop=True)

    plt.plot(case[1])
    plt.savefig(f'case_{num}.png')
    plt.show()

    # case_one 저장
    case.to_csv(f'case_{num}.csv')
```

```python
for i in range(2,5):
    function_concat(i)
```

```python
for i in range(6,8):
    function_concat(i)
```

```python
# 5는 3개까지만 있음
def function_concat_5(num):
    case_temp = []
    # 4로 바꿔줌
    for i in range(1,4):
        case_temp.append(pd.read_csv(f'semina_experiment/{num}-{i}.csv', header=None)[5:])

    case = pd.concat(case_temp).reset_index(drop=True)

    plt.plot(case[1])
    plt.savefig(f'case_{num}.png')
    plt.show()

    # case_one 저장
    case.to_csv(f'case_{num}.csv')
```

```python
function_concat_5(5)
```
