### 함수1

```python
# 기울기 구하는 함수
def diff_search(df_name):
    diff = []
    for i in range(len(df_name)):
        if i == 0:
            diff.append(0)
        else:
            # 데이터프레임 형태 달라지면 이거 바꿔야 함
            diff.append(df_name['1'][i] - df_name['1'][i-1])
    return diff
```

```python
case_1_diff = diff_search(case_1)
```

```python
case_1_diff
```

```python
# 미분값은 무조건 -에서 시작해 +으로 변한다

adfqef = []

for i, diff in enumerate(case_1_diff):
    if -4 < diff < 4:
        continue
    elif diff >= 4:
        adfqef.append(i)
    elif diff <= -4:
        adfqef.append(i)
```
