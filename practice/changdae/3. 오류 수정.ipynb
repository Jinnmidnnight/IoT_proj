```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 파일 한 번에 불러오기
# global 변수 사용하기
for i in range(1,8):
    globals()[f'case_{i}'] = pd.read_csv(f'case_{i}.csv', index_col=0)
```

```python
 import pickle
 list_all = []

for i in range(1,8):
   with open(f'case_{i}_fluc.pkl','rb') as f:
      list_all.append(pickle.load(f))
```
