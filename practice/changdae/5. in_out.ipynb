```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
In_all = pd.read_csv('In_list.csv',header=None,skiprows=[0])
Out_all = pd.read_csv('Out_list.csv',header=None,skiprows=[0])
```

```python
In_all.head()
```

```python
Out_all.head()
```

```python
for i in range(1,70):
    plt.plot(In_all[i])
```

```python
for i in range(1,70):
    plt.figure(figsize=(1,1))
    plt.title(f'in {i} graph')
    plt.plot(In_all[i])
    plt.show()
```

```python
for i in range(1,71):
    plt.figure(figsize=(1,1))
    plt.title(f'out {i} graph')
    plt.plot(Out_all[i])
    plt.show()
```

### 이상한 그래프
* in 20 22
* out 23 24

```python
import random
```

```python
# 임의로 10개의 리스트를 뽑아서 in/out 그래프 그려보기
random.seed(0)
random_list = []
for i in range(10):
    random_list.append(random.randint(0,70))
```

```python
for i in random_list:
    plt.title('10 random In graph')
    plt.plot(In_all[i])
```

```python
for i in random_list:
    plt.title('10 random Out graph')
    plt.plot(Out_all[i])
```
