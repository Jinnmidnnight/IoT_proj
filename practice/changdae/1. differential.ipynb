## 미분 구하는 함수

```python
# 함수로 만들기
def diff_search(df_name):
    diff = []
    for i in range(len(df_name)):
        if i == 0:
            diff.append(0)
        else:
            diff.append(df_name['1'][i] - df_name['1'][i-1])
    return diff

def diff_graph(diff):
    plt.plot(diff)
    # plt.show()

def plot_graph(file_name):
    plt.plot(file_name['1'])
    # plt.show()
```

```python
print('case_1')
plt.subplot(2, 1, 1)
plot_graph(case_1)
plt.subplot(2, 1, 2)
diff_graph(diff_search(case_1))
plt.savefig('case_1_diff.png')
```
